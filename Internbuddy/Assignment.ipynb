## Problem Statment:
InternBuddy (https://www.internbuddy.in) is offering
career opportunity for Budding Data Scientist. Your assignment is to build a
Machine Learning model , which will help the InternBuddy management to
shortlist the potential candidates while applying.  What it means is, your
Machine Learning model needs to be built on given Data-Set

```python
import pandas as pd
from sklearn import linear_model
import math
```

```python
df=pd.read_excel("Data_Science_2020_v2.xlsx")
```

```python
df.describe()
```

```python
df.head()
```

```python
#calculating year of PG
df['Year Of PG']=df['Current Year Of Graduation']+2

#creating new columun for total score and instializing it with 0
df['Score']=0
```

```python
#applying shortlisting criteria

df.loc[df['Current Year Of Graduation']==2020,'Score']=df['Score']+10
df.loc[df['Current Year Of Graduation']==2019,'Score']=df['Score']+8
df.loc[df['Current Year Of Graduation']<=2018,'Score']=df['Score']+5
```

```python
df.loc[df['Year Of PG']==2020,'Score']=df['Score']+7
df.loc[df['Year Of PG']<=2019,'Score']=df['Score']+3
```

```python
df.loc[df['Python (out of 3)']==3,'Score']=df['Score']+10
df.loc[df['Python (out of 3)']==2,'Score']=df['Score']+7
df.loc[df['Python (out of 3)']==1,'Score']=df['Score']+3
```

```python
df.loc[df['R Programming (out of 3)']==3,'Score']=df['Score']+10
df.loc[df['R Programming (out of 3)']==2,'Score']=df['Score']+7
df.loc[df['R Programming (out of 3)']==1,'Score']=df['Score']+3
```

```python
df.loc[df['Data Science (out of 3)']==3,'Score']=df['Score']+10
df.loc[df['Data Science (out of 3)']==2,'Score']=df['Score']+7
df.loc[df['Data Science (out of 3)']==1,'Score']=df['Score']+3
```

```python
df.loc[df['Other skills'].str.find('Machine Learning')>0,'Score']=df['Score']+3
df.loc[df['Other skills'].str.find('Deep Learning')>0,'Score']=df['Score']+3
df.loc[df['Other skills'].str.find('NLP')>0,'Score']=df['Score']+3
df.loc[df['Other skills'].str.find('Statistical Data Analysis')>0,'Score']=df['Score']+3
df.loc[df['Other skills'].str.find('AWS')>0,'Score']=df['Score']+3
df.loc[df['Other skills'].str.find('SQL')+ df['Other skills'].str.find('NoSQL')>0 ,'Score']=df['Score']+3
df.loc[df['Other skills'].str.find('Excel')>0,'Score']=df['Score']+3
```

```python
print('Number of candidates who scored >=40 is',df[df.Score >= 40].shape[0])
```

```python
df.loc[df['Score']>=40,'Shortlisted']='Yes'
df.loc[df['Score']<40,'Shortlisted']='No'
```
