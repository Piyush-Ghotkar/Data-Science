## Problem Statment:

#### InternBuddy (https://www.internbuddy.in) is offering career opportunity for its various project(s). The available job profiles are (a) Web Development and (b) Data Scientist. There are almost 500 candidates applied for the respective role. Your assignment is to build a Machine Learning model , which will help the InternBuddy management to Identify the potential candidates whom they can offer job (to process further for interview). 

```python
import pandas as pd
```

```python
df=pd.read_excel("mavoix_ml_sample_dataset.xlsx")
```

```python
df.describe()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=1,how='all')
```

### Transforming the Performances of 10,12,UG,PG

```python
df['Performance_10']=df['Performance_10'].astype(str)
df['Performance_10']=df['Performance_10'].str.split("/")
df['cgpa']=df['Performance_10'].str.get(0)
df['cgpa']=df['cgpa'].astype('float')

#converting 10 CGPA to 100 percentage
df.loc[df['cgpa']>10,'Performance_10_std']=df['cgpa']
df.loc[df['cgpa']<11,'Performance_10_std']=df['cgpa']*10


#filling na using means
df['Performance_10_std']=df['Performance_10_std'].fillna(df['Performance_10_std'].mean())
```

```python
df['Performance_12']=df['Performance_12'].astype(str)
df['Performance_12']=df['Performance_12'].str.split("/")
df['cgpa']=df['Performance_12'].str.get(0)
df['cgpa']=df['cgpa'].astype('float')

#converting 10 CGPA to 100 percentage
df.loc[df['cgpa']>10,'Performance_12_std']=df['cgpa']
df.loc[df['cgpa']<11,'Performance_12_std']=df['cgpa']*10


#filling na using means
df['Performance_12_std']=df['Performance_12_std'].fillna(df['Performance_12_std'].mean())
```

```python
df['Performance_UG']=df['Performance_UG'].astype(str)
df['Performance_UG']=df['Performance_UG'].str.split("/")
df['cgpa']=df['Performance_UG'].str.get(0)
df['cgpa']=df['cgpa'].astype('float')

#converting 10 CGPA to 100 percentage
df.loc[df['cgpa']>10,'Performance_UG_std']=df['cgpa']
df.loc[df['cgpa']<11,'Performance_UG_std']=df['cgpa']*10


#filling na using means
df['Performance_UG_std']=df['Performance_UG_std'].fillna(df['Performance_UG_std'].mean())
```

```python
df['Performance_PG']=df['Performance_PG'].astype(str)
df['Performance_PG']=df['Performance_PG'].str.split("/")
df['cgpa']=df['Performance_PG'].str.get(0)
df['cgpa']=df['cgpa'].astype('float')

#converting 10 CGPA to 100 percentage
df.loc[df['cgpa']>10,'Performance_PG_std']=df['cgpa']
df.loc[df['cgpa']<11,'Performance_PG_std']=df['cgpa']*10


#filling na using means
df['Performance_PG_std']=df['Performance_PG_std'].fillna(df['Performance_PG_std'].mean())

df=df.drop('cgpa',1)
```

```python
df['Offer_Job']="No"
```

### Now setting thresholds for Web Development Job profile
